## 2. Variables y tipos de datos

En Python no necesitas declarar el tipo de variable, se asigna automáticamente.

Se puede trabajar con datos de tipo numérico o tipo "cadena de texto"

**Tipos Numéricos:**

Son los bloques básicos para cualquier cálculo matemático

`int` **(Enteros):** Números sin decimales, ya sean positivos o negativos (ej. 5, -10)

`float` **(Punto flotante):** Números con decimales (ej. 3.14, 2.0)

`complex` **(Complejos):** Se usan en ingeniería y matemáticas avanzadas. Tienen una parte real y una imaginaria, representada por una $j$ (ej. $2 + 3j$).

**Tipos de Texto:**

Solo hay uno, pero es extremadamente versátil.

`str` **(Strings/Cadenas):** Texto encerrado en comillas simples ('...') o dobles ("..."). Python trata el texto como una secuencia de caracteres.

**Tipo Booleano:**

`bool` **(Booleano):** Solo puede tener dos valores: `True` o `False`. Se usa para representar condiciones lógicas (ej. `es_mayor = True`).

**Tipos de Secuencias:**

| Tipo | ¿Es mutable? | Definición |
|------|:---:|------|
| `list` | Sí | Colección ordenada. Se define con corchetes: `[1, "hola", 3.4]`. |
| `tuple` | No | Colección ordenada inmutable. Se define con paréntesis: `(1, 2, 3)`. |
| `range` | No | Genera una secuencia de números, común en bucles `for`. |
| `dict` | Sí | Colección de pares clave-valor. Se define con llaves: `{"nombre": "Ana", "edad": 25}`. |

```python
# Variables numéricas
edad = 25
estatura = 1.75
print(f"Edad: {edad}, tipo: {type(edad)}")
print(f"Estatura: {estatura}, tipo: {type(estatura)}")

# Variables de texto (strings)
nombre = "Natalia"
print(f"Nombre: {nombre}, tipo: {type(nombre)}")

# Variables booleanas
es_estudiante = True
print(f"¿Es estudiante?: {es_estudiante}, tipo: {type(es_estudiante)}")
```

## 3. Operaciones aritméticas

Python puede usarse como una calculadora.

```python
a = 10
b = 3

print(f"Suma: {a} + {b} = {a + b}")
print(f"Resta: {a} - {b} = {a - b}")
print(f"Multiplicación: {a} * {b} = {a * b}")
print(f"División: {a} / {b} = {a / b:.2f}")
print(f"División entera: {a} // {b} = {a // b}")
print(f"Módulo (residuo): {a} % {b} = {a % b}")
print(f"Potencia: {a} ** {b} = {a ** b}")
```

## 4. Cadenas de texto (Strings)

Las cadenas se pueden manipular de muchas formas.

```python
saludo = "Hola, bienvenido al curso"

# Métodos básicos de strings
print(saludo.upper())          # Todo en mayúsculas
print(saludo.lower())          # Todo en minúsculas
print(saludo.replace("curso", "taller"))  # Reemplazar texto
print(len(saludo))             # Longitud de la cadena

# Concatenar strings
nombre = "Natalia"
mensaje = saludo + ", " + nombre + "!"
print(mensaje)

# f-strings (formato moderno)
print(f"{saludo}, {nombre}!")
```

## 5. Listas

Las listas son colecciones ordenadas que pueden contener cualquier tipo de dato.

```python
# Crear una lista
frutas = ["manzana", "banana", "cereza", "durazno"]
print(f"Lista: {frutas}")

# Acceder a elementos (el índice empieza en 0)
print(f"Primera fruta: {frutas[0]}")
print(f"Última fruta: {frutas[-1]}")

# Agregar y eliminar elementos
frutas.append("uva")
print(f"Después de agregar: {frutas}")

frutas.remove("banana")
print(f"Después de eliminar: {frutas}")

# Longitud de la lista
print(f"Total de frutas: {len(frutas)}")

# Rebanar (slicing)
print(f"Primeras dos: {frutas[:2]}")
```

## 6. Condicionales (if / elif / else)

Permiten ejecutar código según una condición.

```python
nota = 85

if nota >= 90:
    print("Excelente")
elif nota >= 70:
    print("Aprobado")
else:
    print("Reprobado")

# Operadores de comparación
print(f"10 > 5: {10 > 5}")
print(f"10 == 10: {10 == 10}")
print(f"10 != 5: {10 != 5}")
```

## 7. Ciclos (Loops)

### Ciclo `for` — recorre una secuencia

```python
# Recorrer una lista
colores = ["rojo", "verde", "azul"]
for color in colores:
    print(f"Color: {color}")

print("---")

# Usar range para generar números
for i in range(1, 6):
    print(f"Número: {i}")
```

### Ciclo `while` — repite mientras una condición sea verdadera

```python
contador = 1
while contador <= 5:
    print(f"Contando: {contador}")
    contador += 1  # Incrementar el contador

print("¡Fin del conteo!")
```

### Solución

```python
# Lista de notas
notas = [4.5, 3.2, 1.8, 2.5, 4.8]

# Calcular el promedio
suma = sum(notas)
cantidad = len(notas)
promedio = suma / cantidad

print(f"Notas: {notas}")
print(f"Suma: {suma}")
print(f"Cantidad de estudiantes: {cantidad}")
print(f"Promedio: {promedio:.2f}")

# Determinar si el grupo aprobó
if promedio >= 70:
    print("El grupo aprobó en promedio")
else:
    print("El grupo reprobó en promedio")
```
