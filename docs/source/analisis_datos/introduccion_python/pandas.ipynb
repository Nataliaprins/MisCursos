## 1. ¿Qué es Pandas?

**Pandas** es una librería de Python diseñada para la manipulación y el análisis de datos estructurados (tablas, hojas de cálculo, bases de datos, etc.).

Sus dos estructuras principales son:

| Estructura | Descripción |
|---|---|
| `Series` | Una columna de datos (como un vector o una lista con etiquetas). |
| `DataFrame` | Una tabla completa con filas y columnas (como una hoja de Excel). |

Para usarla, primero hay que importarla:

```python
import pandas as pd  # "pd" es la convención estándar
print(f"Versión de Pandas: {pd.__version__}")
```

## 2. Series y DataFrames

### Series
Una `Series` es como una lista con índices (etiquetas).

```python
# Crear una Serie desde una lista
notas = pd.Series([85, 92, 78, 95, 88], name="Notas")
print(notas)
print(f"\nTipo: {type(notas)}")

# Crear una Serie con índices personalizados
notas_estudiantes = pd.Series(
    [85, 92, 78, 95, 88],
    index=["Ana", "Luis", "María", "Carlos", "Sofía"],
    name="Notas"
)
print("\nNotas con nombres:")
print(notas_estudiantes)

# Acceder a un valor por su índice
print(f"\nNota de Ana: {notas_estudiantes['Ana']}")
```

### DataFrame
Un `DataFrame` es una tabla con filas y columnas. Se puede crear desde un diccionario.

```python
# Crear un DataFrame desde un diccionario
datos = {
    "Nombre": ["Ana", "Luis", "María", "Carlos", "Sofía"],
    "Edad": [21, 23, 22, 24, 20],
    "Nota": [85, 92, 78, 95, 88],
    "Ciudad": ["Bogotá", "Medellín", "Cali", "Bogotá", "Medellín"]
}

df = pd.DataFrame(datos)
print("DataFrame:")
df
```

## 3. Lectura de datos desde archivos

Pandas puede leer datos de múltiples formatos:

| Función | Formato |
|---|---|
| `pd.read_csv()` | Archivos CSV |
| `pd.read_excel()` | Archivos Excel (.xlsx) |
| `pd.read_json()` | Archivos JSON |
| `pd.read_sql()` | Bases de datos SQL |

Para este ejemplo, guardaremos nuestro DataFrame como CSV y luego lo leeremos.

```python
# Guardar el DataFrame como archivo CSV
df.to_csv("estudiantes.csv", index=False)
print("Archivo 'estudiantes.csv' guardado.")

# Leer el archivo CSV
df_leido = pd.read_csv("estudiantes.csv")
print("\nDataFrame leído desde CSV:")
df_leido
```

## 4. Exploración de datos

Cuando recibes datos nuevos, lo primero es explorarlos para entender su estructura.

```python
# Ver las primeras filas
print("Primeras 3 filas:")
print(df.head(3))

# Ver las últimas filas
print("\nÚltimas 2 filas:")
print(df.tail(2))

# Dimensiones del DataFrame (filas, columnas)
print(f"\nDimensiones: {df.shape}")
print(f"Filas: {df.shape[0]}, Columnas: {df.shape[1]}")

# Nombres de las columnas
print(f"\nColumnas: {list(df.columns)}")

# Tipos de datos de cada columna
print("\nTipos de datos:")
print(df.dtypes)
```

```python
# Resumen de información general
print("Información del DataFrame:")
print(df.info())

# Estadísticas descriptivas de columnas numéricas
print("\nEstadísticas descriptivas:")
df.describe()
```

## 5. Selección y filtrado de datos

### Seleccionar columnas

```python
# Seleccionar una sola columna (devuelve una Serie)
print("Columna 'Nombre':")
print(df["Nombre"])

# Seleccionar varias columnas (devuelve un DataFrame)
print("\nColumnas 'Nombre' y 'Nota':")
df[["Nombre", "Nota"]]
```

### Seleccionar filas con `iloc` y `loc`

- `iloc` → selección por **posición** (índice numérico)
- `loc` → selección por **etiqueta** (nombre del índice)

```python
# Seleccionar la primera fila por posición
print("Primera fila (iloc):")
print(df.iloc[0])

# Seleccionar filas 1 a 3
print("\nFilas 1 a 3:")
print(df.iloc[1:4])

# Seleccionar fila y columna específica
print(f"\nNombre de la fila 2: {df.iloc[2, 0]}")
print(f"Nota de la fila 2: {df.iloc[2, 2]}")
```

### Filtrar datos con condiciones

Puedes filtrar filas que cumplan una condición lógica.

```python
# Estudiantes con nota mayor o igual a 90
print("Estudiantes con nota >= 90:")
aprobados = df[df["Nota"] >= 90]
print(aprobados)

# Estudiantes de Bogotá
print("\nEstudiantes de Bogotá:")
bogota = df[df["Ciudad"] == "Bogotá"]
print(bogota)

# Combinar condiciones: nota >= 85 Y edad < 23
print("\nNota >= 85 y edad < 23:")
filtro = df[(df["Nota"] >= 85) & (df["Edad"] < 23)]
print(filtro)
```

## 6. Operaciones básicas

### Agregar y eliminar columnas

```python
# Agregar una nueva columna calculada
df["Aprobado"] = df["Nota"] >= 70
print("Con columna 'Aprobado':")
print(df)

# Agregar una columna con operación aritmética
df["Nota_sobre_100"] = df["Nota"] / 100
print("\nCon columna 'Nota_sobre_100':")
print(df)

# Eliminar una columna
df = df.drop(columns=["Nota_sobre_100"])
print("\nSin 'Nota_sobre_100':")
print(df)
```

### Ordenar datos

```python
# Ordenar por nota (ascendente)
print("Ordenado por nota (menor a mayor):")
print(df.sort_values("Nota"))

# Ordenar por nota (descendente)
print("\nOrdenado por nota (mayor a menor):")
print(df.sort_values("Nota", ascending=False))
```

### Agrupar datos con `groupby`

Permite agrupar filas por una columna y calcular estadísticas por grupo.

```python
# Promedio de nota por ciudad
print("Promedio de nota por ciudad:")
print(df.groupby("Ciudad")["Nota"].mean())

# Contar estudiantes por ciudad
print("\nEstudiantes por ciudad:")
print(df.groupby("Ciudad")["Nombre"].count())

# Múltiples estadísticas por grupo
print("\nEstadísticas de nota por ciudad:")
print(df.groupby("Ciudad")["Nota"].agg(["mean", "min", "max"]))
```

3. ¿Cuál es el total de ventas por producto?

```python
# Datos de ejemplo
ventas = pd.DataFrame({
    "Producto": ["Laptop", "Mouse", "Teclado", "Monitor", "Laptop", "Mouse", "Monitor"],
    "Cantidad": [2, 10, 5, 3, 1, 8, 2],
    "Precio_unitario": [800, 25, 45, 350, 800, 25, 350],
    "Vendedor": ["Pedro", "Ana", "Ana", "Pedro", "Ana", "Pedro", "Ana"]
})

# Agregar columna de venta total
ventas["Venta_total"] = ventas["Cantidad"] * ventas["Precio_unitario"]
ventas
```

### Solución

```python
# 1. Dimensiones del DataFrame
print(f"Dimensiones: {ventas.shape} → {ventas.shape[0]} filas, {ventas.shape[1]} columnas")

# 2. Ventas mayores a $500
print("\nVentas mayores a $500:")
print(ventas[ventas["Venta_total"] > 500])

# 3. Total de ventas por producto
print("\nTotal de ventas por producto:")
print(ventas.groupby("Producto")["Venta_total"].sum())

# 4. Ordenar por venta total (mayor a menor)
print("\nOrdenado por venta total:")
print(ventas.sort_values("Venta_total", ascending=False))
```
