# 2. Principios básicos para la construcción de portafolios.

Revisa los conceptos básicos de portafolios de inversión. [Descargar: Fundamentos de carteras](2_Fundamentos_carteras.pdf)


Supongamos 3 activos (A, B, C) y los rendimientos son los siguientes:

```python
retornos_A = [0.05, -0.02, 0.08, -0.04, 0.03]
retornos_B = [0.02, 0.01, 0.03, -0.01, 0.02]
retornos_C = [0.01, 0.005, 0.01, 0.008, 0.012]
```

Calculemos la media:


```python
rdto_media_A= sum(retornos_A) / len(retornos_A)
rdto_media_B= sum(retornos_B) / len(retornos_B)
rdto_media_C= sum(retornos_C) / len(retornos_C)
print(f"Rendimiento medio de A: {rdto_media_A:.4f}")
print(f"Rendimiento medio de B: {rdto_media_B:.4f}")
print(f"Rendimiento medio de C: {rdto_media_C:.4f}")
```

Calculemos la varianza de los activos: 

```python
var_A = sum((x - rdto_media_A) ** 2 for x in retornos_A) / (len(retornos_A) - 1)
var_B = sum((x - rdto_media_B) ** 2 for x in retornos_B) / (len(retornos_B) - 1)
var_C = sum((x - rdto_media_C) ** 2 for x in retornos_C) / (len(retornos_C) - 1)
print(f"Varianza de A: {var_A:.6f}")
print(f"Varianza de B: {var_B:.6f}")
print(f"Varianza de C: {var_C:.6f}")    
```

Calculemos la covarianza de (A,B)

```python
def calcular_covarianza(datos_x, media_x, datos_y, media_y):
    suma_productos = sum((x - media_x) * (y - media_y) for x, y in zip(datos_x, datos_y))
    return suma_productos / (len(datos_x) - 1)

cov_AB = calcular_covarianza(retornos_A, rdto_media_A, retornos_B, rdto_media_B)
cov_AC = calcular_covarianza(retornos_A, rdto_media_A, retornos_C, rdto_media_C)
cov_BC = calcular_covarianza(retornos_B, rdto_media_B, retornos_C, rdto_media_C)
print(f"Covarianza entre A y B: {cov_AB:.6f}")
print(f"Covarianza entre A y C: {cov_AC:.6f}")
print(f"Covarianza entre B y C: {cov_BC:.6f}")
```

Calculemos el rendiemiento medio del portafolio con 2 activos:

```python
# Veamos que pasa con una diversificación ingenua con solo los activos A y B, con pesos iguales
peso_A = 0.5
peso_B = 0.5

# Rendimiento esperado del portafolio
rdto_portafolio = peso_A * rdto_media_A + peso_B * rdto_media_B
print(f"Rendimiento esperado del portafolio: {rdto_portafolio:.4f}")    

```

Ahora, calculemos la varianza del portafolio:

```python
# Cálculo de la varianza del portafolio con los activos A y B
var_portafolio = (peso_A ** 2) * var_A + (peso_B ** 2) * var_B + 2 * peso_A * peso_B * cov_AB
print(f"Varianza del portafolio: {var_portafolio:.6f}")
```

Recordemos que una forma de hallar volatilidad es calculando la desviación estandar, que a su vez, es la raiz de la varianza, por tanto la volatilidad del portafolio es:


```python
vol_portafolio = var_portafolio ** 0.5
print(f"Volatilidad del portafolio: {vol_portafolio:.6f}")
```

**¿Que pasa si agregamos un activo mas?**

Construyamos una cartera ingenua con 3 activos



```python
# portafolio con los tres activos y pesos iguales
peso_A = 1/3
peso_B = 1/3
peso_C = 1/3

rdto_portafolio_3 = peso_A * rdto_media_A + peso_B * rdto_media_B + peso_C * rdto_media_C
print(f"Rendimiento esperado del portafolio con 3 activos: {rdto_portafolio_3:.4f}")
```

```python
#varianza del portafolio con los tres activos
var_portafolio_3 = (peso_A ** 2) * var_A + (peso_B ** 2) * var_B + (peso_C ** 2) * var_C
var_portafolio_3 += 2 * peso_A * peso_B * cov_AB + 2 * peso_A * peso_C * cov_AC + 2 * peso_B * peso_C * cov_BC
print(f"Varianza del portafolio con 3 activos: {var_portafolio_3:.6f}") 
```

Construyamos la matriz  de covarianzas

```python
#contruyamos la matriz de varianzas y covarianzas
matriz_covarianzas = [
    [var_A, cov_AB, cov_AC],
    [cov_AB, var_B, cov_BC],
    [cov_AC, cov_BC, var_C]
]   
print("Matriz de varianzas y covarianzas:")
for fila in matriz_covarianzas:
    print(fila) 
```

```python
# calculamos la volatilidad del portafolio con los tres activos
vol_portafolio_3 = var_portafolio_3 ** 0.5
print(f"Volatilidad del portafolio con 3 activos: {vol_portafolio_3:.6f}")      
```
