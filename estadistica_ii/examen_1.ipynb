```python
import numpy as np
import pandas as pd

# Parámetros de la simulación
n_sur = 70
n_norte = 120

# Supongamos medias y desviaciones estándar distintas para cada zona
media_sur = 35  # minutos
std_sur = 8
media_norte = 40  # minutos
std_norte = 10

# Simulación de tiempos de entrega
np.random.seed(42)  # reproducibilidad

tiempo_sur = np.random.normal(loc=media_sur, scale=std_sur, size=n_sur)
tiempo_norte = np.random.normal(loc=media_norte, scale=std_norte, size=n_norte)

# Crear DataFrame
entregas = pd.DataFrame({
    'zona': ['Sur'] * n_sur + ['Norte'] * n_norte,
    'tiempo_entrega': np.concatenate([tiempo_sur, tiempo_norte])
})

entregas.head()
```

```python
entregas.to_csv("entregas.csv", index=False)
```
